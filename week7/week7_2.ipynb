```python
plt.figure()
plt.plot(x, x ** 2, 'r-+')
plt.plot(x, x ** 3, 'b:*')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Graph of $x$ versus $x^2$ and $x^3$")
plt.xlim(1,3)
plt.ylim(-5,15) # y   
```

```python
plt.figure()
plt.plot(x, x**2, 'r-+')
plt.plot(x, x **3, 'b:*')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Graph of $x$ versus $x^2$ and $x^3$")
plt.legend(['$x^2$','$x^3$'], loc = 'upper center')
```

```python
fig = plt.figure()
plt.plot(x, x**2, 'r-+')
plt.plot(x, x**3, 'b:*')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Graph of $x$ versus $x^.2$ and $x^3$")
plt.legend(['$x^2$', '$x^3$'], loc = "upper left")
fig.savefig('graph.png')
```

```python
fig, axes = plt.subplots(nrows= 2, ncols = 3)

for row in range(2):
    for col in range(3):
        axe = axes[row][col]
        axe.plot(x,x**2)
        axe.set_xlabel('x')
        axe.set_ylabel('y')
        axe.set_title('x v.s. "$x^2$"')
        axe.set_xlim((0,4))
        axe.set_ylim((0,20))
    fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 3 )
for row in range(2):
    for col in range(3):
        axe = axes[row][col]
        axe.plot(x, x**2)
        axe.set_xlabel('x')
        axe.set_ylabel('y')
        axe.set_title('x v.s. $x^2$')
        axe.set_xlim((0,4))
        axe.set_ylim((0,20))
fig.subplots_adjust(wspace=0.3, hspace=0.5) # not fig.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2, figsize = (12,4))

axes[0].plot(x, x**2, x, x,)
axes[0].set_title("linear scale")

axes[1].plot(x, x**2, x, x)
axes[1].set_title("log scale")
axes[1].set_yscale('log', base = 2)
```

```python
x1 = np.linspace(-0.75, 1.0, 100)
n = np.array(range(6))
fig, axes = plt.subplots(nrows = 1, ncols = 4, figsize = (20,5))

axes[0].scatter(x1, x1 + 0.2 * np.random.randn(len(x1)))
axes[1].set_title('scatter')

axes[1].step(n, n ** n, where = 'mid')
axes[1].set_title('step')

axes[2].bar(n, n** 2, align = 'center', width = 0.6)
axes[2].set_title('bar')

axes[3].fill_between(x, x**2, x**3, color = 'red')
axes[3].set_title('fill_between')

fig.tight_layout()
```

```python
data = np.random.lognormal(size = (37,4), mean = 1.5, sigma = 1.75)
labels = list('ABCD')
fig, axes = plt.subplots(nrows = 2, ncols = 2, figsize = (16,6))
axes[0, 0].boxplot(data, labels = labels)
axes[0, 0].set_title('Default')
axes[0, 1].boxplot(data, labels = labels, showmeans = True)
axes[0, 1].set_title("showmean=True")
axes[1, 0].boxplot(data, labels = labels, showmeans = True, meanline = True)
axes[1,0].set_title("showmeans = True\mealine=True")
axes[1, 1].boxplot(data, labels = labels, showbox = False, showcaps = False)
axes[1, 1].set_title("showmeans=False\showcaps=False")
axes[0, 0].set_yscale('log')
axes[0, 0].set_yticklabels([])
axes[0, 1].set_yscale('log')
axes[0, 1].set_yticklabels([])
axes[1, 0].set_yscale('log')
axes[1, 0].set_yticklabels([])
axes[1, 1].set_yscale('log')
axes[1, 1].set_yticklabels([])
```
