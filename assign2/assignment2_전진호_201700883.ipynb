```python
from sklearn.datasets import load_iris
from sklearn.model_selection import StratifiedKFold
iris = load_iris()
X, y = iris.data, iris.target
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
sc = StandardScaler()
```

```python
X_scaled = sc.fit_transform(X)
```

```python
n_splits = 10
kf = StratifiedKFold(n_splits, shuffle = True)

total = []
for k in range(1,31):
    fold_scores = []
    for train_index, test_index in kf.split(X_scaled, y):
        X_train, X_test = X_scaled[train_index], X_scaled[test_index]
        Y_train, Y_test = y[train_index], y[test_index]
        knn = KNeighborsClassifier(k)
        knn.fit(X_train, Y_train)
        Y_pred = knn.predict(X_test)
        
        predicted_accuracy = accuracy_score(Y_test, Y_pred)
        fold_scores.append(predicted_accuracy)
    
    total.append(np.mean(fold_scores))

optimal_k = np.argmax(total)
print("10-fold KNN becomes optimal when k =", optimal_k)
print("optimal_scores = ", total[optimal_k])
```

```python
MSE_list = []

for deg in range(1,11):
    poly = sklearn.preprocessing.PolynomialFeatures(degree=deg)
    X_poly_train = poly.fit_transform(X_train)
    X_poly_test = poly.transform(X_test)
    lr = sklearn.linear_model.LinearRegression()
    lr.fit(X_poly_train, Y_train)
#     print(X_poly_train.shape, Y_train.shape)
    Y_pred = lr.predict(X_poly_test)
    error = sklearn.metrics.mean_squared_error(Y_test, Y_pred)
    MSE_list.append(error)



optimal_index = np.argmin(MSE_list)+1
print("optimal_index", optimal_index)
print("optimal_eror", MSE_list[optimal_index])
    
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,11), np.log10(MSE_list), 'bo--')
plt.ylabel('MSE with log10')
plt.xlabel('degree from n=1 to 10')
plt.grid()
```
